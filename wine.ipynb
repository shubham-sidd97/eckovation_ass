```python
dict={}
for i in range(len(wine)):
    if(wine['quality'][i] not in dict):
        dict[wine['quality'][i]]=[[],[]]
    dict[wine['quality'][i]][0].append(wine['fixed.acidity'][i])
    dict[wine['quality'][i]][1].append(wine['volatile.acidity'][i])
```

```python
import numpy as np

import matplotlib.pyplot as plt
poor=[[],[]]
avg=[[],[]]
best=[[],[]]
for i in range(2):
    poor[0]=dict[3][0]+dict[4][0]
    poor[1]=dict[3][1]+dict[4][1]
    avg[0]=dict[5][0]+dict[6][0]
    avg[1]=dict[5][1]+dict[6][1]
    best[0]=dict[7][0]+dict[8][0]+dict[9][0]
    best[1]=dict[7][1]+dict[8][1]+dict[9][1]
```

```python
a=poor[0]
b=poor[1]
c=avg[0]
d=avg[1]
e=best[0]
f=best[1]
plt.scatter(a,b)
plt.scatter(c,d)
plt.scatter(e,f)

plt.xlabel('fixed.acidity')
plt.ylabel('volatile.acidity')
```

```python
import json
x=0.0
y=0.0
z=0.0
i=0
j=0
rank=0
dic={}
diccl={}
diccd={}
diccn={}
dicn={}
idlis=[]
lis=[0,0,0,0,0]
```

```python
f1 = open("file1.json").read()
f1 = json.loads(f1)
f2 = open("file2.json").read()
f2 = json.loads(f2)
f3 = open("file3.json").read()
f3 = json.loads(f3)
f4 = open("file4.json").read()
f4 = json.loads(f4)
f5 = open("file5.json").read()
f5 = json.loads(f5)
files=[f1,f2,f3,f4,f5]
```

```python
for d in files:
    for i in range(len(d)):
        if("restaurants" in d[i]):
            for j in range(len(d[i]["restaurants"])):
                if (d[i]["restaurants"][j]["restaurant"]["id"] not in idlis):
                    if(d[i]["restaurants"][j]["restaurant"]["location"]["city_id"] not in diccl):
                        diccl[d[i]["restaurants"][j]["restaurant"]["location"]["city_id"]]=[0,0,0,0,0]
                        diccd[d[i]["restaurants"][j]["restaurant"]["location"]["city_id"]]={}
                        dicn[d[i]["restaurants"][j]["restaurant"]["location"]["city_id"]]=d[i]["restaurants"][j]["restaurant"]["location"]["city"]
                    lis=diccl[d[i]["restaurants"][j]["restaurant"]["location"]["city_id"]]
                    dic=diccd[d[i]["restaurants"][j]["restaurant"]["location"]["city_id"]]
                    x=float(d[i]["restaurants"][j]["restaurant"]["user_rating"]["votes"])
                    y=float(d[i]["restaurants"][j]["restaurant"]["user_rating"]["aggregate_rating"])
                    idlis.append(d[i]["restaurants"][j]["restaurant"]["id"])
                    if(lis[0]<x*y):
                        if(len(dic)==5):
                            del dic[lis[0]]
                        lis[0]=x*y
                        lis.sort()
                        dic[x*y]=d[i]["restaurants"][j]["restaurant"]["name"]  
                    diccl[d[i]["restaurants"][j]["restaurant"]["location"]["city_id"]]=lis
                    diccd[d[i]["restaurants"][j]["restaurant"]["location"]["city_id"]]=dic
                    diccn[x*y]=d[i]["restaurants"][j]["restaurant"]["location"]["address"]
```

```python
for city_id in diccl:
    diccl[city_id].reverse()
    print("\nCity ID = ",city_id)
    print("City Name = ",dicn[city_id])
    for rank in range(5):
        print("Rank = ",rank+1)
        print("Name = ",diccd[city_id][diccl[city_id][rank]])
        print("Address = ",diccn[diccl[city_id][rank]])
```
